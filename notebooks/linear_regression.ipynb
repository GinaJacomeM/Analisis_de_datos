```python
features = [col for col in df_car_data.select_dtypes(include='number').columns if col != 'Selling_Price']

# Función para analizar y graficar outliers usando el método del IQR
def analize_outliers(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[columna] < lower_bound) | (df[columna] > upper_bound)]
    
    print(f"\nEstadísticas para {columna}:")
    print(f"Q1 - 1.5IQR = {lower_bound:.2f}")
    print(f"Q1 = {Q1:.2f}")
    print(f"Mediana = {df[columna].median():.2f}")
    print(f"Q3 = {Q3:.2f}")
    print(f"Q3 + 1.5IQR = {upper_bound:.2f}")
    print(f"Número de observaciones: {len(df[columna])}")
    print(f"Número de outliers: {len(outliers)}")

    # Gráfico de caja (boxplot individual)
    plt.figure(figsize=(7, 3))
    sns.boxplot(x=df[columna], color='skyblue')
    plt.title(f'Boxplot para {columna}')
    plt.xlabel(columna)
    plt.tight_layout()
    plt.show()
    
    return outliers

# Diccionario para guardar outliers por variable
outliers_dict = {}

for col in features:
    outliers = analize_outliers(df_car_data, col)
    outliers_dict[col] = outliers
```

```python
def remove_outliers_iqr(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[columna] >= lower_bound) & (df[columna] <= upper_bound)]
```

```python
df_clean = df_car_data.copy()

for col in features:
    df_clean = remove_outliers_iqr(df_clean, col)

df_clean

df_clean.info()
```

```python
# convertir variables de objeto a variables categóricas

df_clean['Fuel_Type'] = df_clean['Fuel_Type'].astype('category')

df_clean['Seller_Type'] = df_clean['Seller_Type'].astype('category')

df_clean['Transmission'] = df_clean['Transmission'].astype('category')

df_clean['Owner'] = df_clean['Owner'].astype('category')

print(df_clean.info())
```

```python
# Filtrar solo los datos dentro de los rangos válidos
df_model = df_clean.copy()

df_model.head(10)
```

```python
df_model.info()
```

```python
df_model['Seller_Type'] = df_model['Seller_Type'].map({'Dealer': 0, 'Individual': 1})
df_model['Fuel_Type'] = df_model['Fuel_Type'].map({'Petrol': 0, 'Diesel': 1, 'CNG': 2})
df_model['Transmission'] = df_model['Transmission'].map({'Manual': 0, 'Automatic': 1})

df_model
```

```python
# Transformar la variable objetivo
df_model['selling_price_log'] = np.log1p(df_model['Selling_Price'])

# Crear nueva columna: antigüedad del auto
df_model['car_age'] = 2025 - df_model['Year']

#Se seleccionan las variables predictoras (X) y la objetivo (y)
feature_cols = [
    'car_age',
    'Present_Price',
    'Kms_Driven',
    'Transmission', 
    'Fuel_Type', 
    'Seller_Type',
    'Owner'
]

X = df_model[feature_cols]
y = df_model['selling_price_log']
```

```python

selector = SelectKBest(score_func=f_regression, k='all')
selector.fit(X, y)
mask = selector.pvalues_ < 0.05
selected = X.loc[:, mask]

# Dividir dataset
X_train, X_test, y_train, y_test = train_test_split(selected, y, test_size=0.2, random_state=42)


selected
```

```python
# Entrenar modelo
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# Evaluar
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

print("Métricas del modelo:")
print(f"R² Score: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
```

```python
# Visualizar resultados

residuals = y_test - y_pred
plt.figure(figsize=(8,5))
sns.histplot(residuals, kde=True)
plt.title("Distribución de errores (residuos)")
plt.xlabel("Error (log)")
plt.show()

print("Tamaño del conjunto de prueba:", len(y_test))
print("Primeras predicciones:", y_pred[:5])
print("Primeros valores reales:", y_test[:5].values)
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x=y_test, y=y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
plt.xlabel("Valores Reales")
plt.ylabel("Valores Predichos")
plt.title("Predicción vs Realidad")
plt.show()
```

```python
# Gráfico: Real vs Predicción 
fig, ax1 = plt.subplots(dpi=300)
plt.rcParams["font.family"] = "Times New Roman"  # opcional

ax1.scatter(y_test, y_pred, marker='o', facecolors='none', edgecolors='r', alpha=1)
plt.xlabel('Precio de venta real')
plt.ylabel('Precio de venta previsto')
plt.title('Predicción de regresión lineal')
plt.xlim(left=0, right=25)
plt.ylim(bottom=0, top=25)
ax1.xaxis.set_minor_locator(MultipleLocator(4))
ax1.yaxis.set_minor_locator(MultipleLocator(4))
ax1.tick_params(which='major', length=5)
ax1.tick_params(which='minor', length=2)

# Línea de ajuste perfecto
Perfect_fit = ax1.axline([0, 0], [25, 25], color='black', linewidth=2, label='Ajuste perfecto')

a, b = np.polyfit(y_test, y_pred, 1)

# Crear un rango continuo de valores x para graficar la línea de mejor ajuste
x_line = np.linspace(0, 25, 100)
y_line = a * x_line + b

# Línea de mejor ajuste (línea púrpura)
Linear_Predict, = plt.plot(x_line, y_line, color="purple", linewidth=1,
                           dashes=[0, 2, 5], label="Lineal (Predicho)")

# Agregar leyenda y métricas
ax1.legend(handles=[Perfect_fit, Linear_Predict])
plt.text(17, 10, f'$R^2= {round(r2, 3)}$', fontsize=10)
plt.text(17, 8, f'$RMSE= {round(rmse, 3)}$', fontsize=10)
plt.text(17, 6, f'$MAE= {round(mae, 3)}$', fontsize=10)
plt.show()
```

```python
# Gráfico: Comparación por muestra 
fig, ax = plt.subplots(figsize=(12, 3.5), dpi=300)
plt.rcParams["font.family"] = "Times New Roman"

line1, = ax.plot(range(len(y_test)), y_test, '-b', label='Valor real', marker='.')
line2, = ax.plot(range(len(y_pred)), y_pred, 'r', label='Predicción de regresión lineal',
                 marker='s', linestyle='--', markersize=3.5)

ax.legend(handles=[line1, line2], loc='best')
plt.xlabel('Muestras')
plt.ylabel('Precio de venta')
plt.title("Predicciones vs. Valores Reales - Comparación de Muestras")
plt.show()
```
